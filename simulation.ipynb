```python
#initiera gdirsen
base_url = "https://cluster.klima.uni-bremen.de/~oggm/gdirs/oggm_v1.6/L3-L5_files/2023.3/elev_bands/W5E5_spinup"

gdirs = workflow.init_glacier_directories(
    glacier_selection.RGIId.values, from_prepro_level=5, prepro_border=160, prepro_base_url=base_url
)
```

```python
#initiera klimatdatan
warnings.simplefilter(action='ignore', category=FutureWarning)
member = "mri-esm2-0_r1i1p1f1"
ssps = ["ssp126","ssp370","ssp585"]

#skapa klimatdata för att sedan köra simulationer med
for ssp in ssps:
    # bias correct them
    workflow.execute_entity_task(
        gcm_climate.process_monthly_isimip_data,
        gdirs,
        ssp=ssp,
        # gcm member -> you can choose another one
        member=member,
        # name of climatefile
        output_filesuffix=f"_Klimatdata_{member}_{ssp}",
    );
```

```python
#Simulerar med klimatdatan.

for ssp in ssps:
    # Här definierar vi namnet på datan som vi vill använda.
    rid = f"_Klimatdata_{member}_{ssp}"
    # Funktion som kör modellen.
    workflow.execute_entity_task(
        tasks.run_with_hydro,
        gdirs,
        # Modellen ska köras från klimatdata.
        run_task=tasks.run_from_climate_data,
        ys=1980,
        # use gcm_data, not climate_historical
        climate_filename="gcm_data",
        # use the chosen scenario
        climate_input_filesuffix=rid,
        # this is important! Start from 2020 glacier
        init_model_filesuffix="_spinup_historical",
        # recognize the run for later
        output_filesuffix=rid,
        # add monthly diagnostics
        store_monthly_hydro=True,
    );
```

```python
# Här kombinerar vi de separata filerna till en df som heter compiled_data
rid_ssp126 = f"_Klimatdata_{member}_ssp126."
rid_ssp370 = f"_Klimatdata_{member}_ssp370"
rid_ssp585 = f"_Klimatdata_{member}_ssp585"
for ssp in ssps:
    rid = f"_Klimatdata_{member}_{ssp}"
    compiled_data = utils.compile_run_output(gdirs,
                                         input_filesuffix=rid,
                                        path=f"/home/ostronet/kurser/see110/sim_outputs/output_{ssp}_{member}/")
```
